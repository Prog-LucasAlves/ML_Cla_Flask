## Create fake data to insert into the template form

```python
# Libraries

import random
import pandas as pd
import requests
import time
```

```python
# Loading the dataset

data = pd.read_csv("../data/diabetes_prediction_dataset.csv")

# View the first few rows of the dataset

data.head()
```

```python
for i in range(20000):
    gender = random.choice(["Female", "Male"])
    age = random.randint(18, 80)
    hypertension = random.choice([0, 1])
    heart_disease = random.choice([0, 1])
    smoking_history = random.choice(
        ["never", "No Info", "current", "former", "ever", "not current"]
    )
    bmi = round(random.uniform(10, 96), 2)
    hba1c_level = round(random.uniform(3, 9), 1)
    blood_glucose_level = random.randint(80, 300)

    print(f"Fake data {i + 1}:\n")
    print(f"Gender: {gender}")
    print(f"Age: {age}")
    print(f"Hypertension: {hypertension}")
    print(f"Heart Disease: {heart_disease}")
    print(f"Smoking History: {smoking_history}")
    print(f"BMI: {bmi}")
    print(f"Hba1c Level: {hba1c_level}")
    print(f"Blood Glucose Level: {blood_glucose_level}")

    url = "https://ml-cla-flask.onrender.com/predict"

    fake = {
        "gender": gender,
        "age": age,
        "hypertension": hypertension,
        "heart_disease": heart_disease,
        "smoking_history": smoking_history,
        "bmi": bmi,
        "hba1c_level": hba1c_level,
        "blood_glucose_level": blood_glucose_level,
    }

    headrers = {"Content-Type": "application/json"}
    try:
        response = requests.post(url, json=fake, headers=headrers)

        if response.status_code == 200:
            result = response.json()
            print("\nSucesso")
            print(f"Previsão: {result['prediction']}")
            print(f"Probabilidades: {result.get('probability', 'N/A')}")
            print("####################\n")

        else:
            print(f"Erro: {response.status_code}")
            print(response.json())

    except requests.exceptions.RequestException as e:
        print(f"Erro de conexão: {e}")

    time.sleep(5)
```
